```python
import numpy as np
import pandas as pd

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

import seaborn as sns
```

```python
df = pd.read_csv("train (1).csv")[['Age','Pclass','SibSp','Parch','Survived']]
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
X=df.iloc[:,0:4]
y=df.iloc[:,-1]
```

```python
X.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=20))
```

Applying Feature Construction

```python
 X['Family_size']=X['SibSp']+X['Parch']+1
```

```python
X.head()
```

```python
def myfunc(num):
    if num==1:
      #alone
        return 0
    elif num>1 and num<=4:
      #small family
            return 1
    else:
      #Large family
            return 2
```

```python
myfunc(4)
```

```python
X['Family_type']=X['Family_size'].apply(myfunc)
```

```python
X.head()
```

```python
X.drop(columns=['SibSp','Parch','Family_size'],inplace=True)
```

```python
X.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=20))
```

Feature Splitting

```python
df=pd.read_csv('train (1).csv')
```

```python
df.head()
```

```python
df['Name']
```

```python
df['Title']=df['Name'].str.split(',',expand=True)[1].str.split('.',expand=True)[0]
```

```python
df['Name'].str.split(',',expand=True)[1].str.split('.',expand=True)
```

```python
df[['Title','Name']]
```

```python
# Convert 'Survived' column to numeric before calculating the mean
df['Survived'] = pd.to_numeric(df['Survived'], errors='coerce')

# Extract the title correctly using str.extract
df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)

# Calculate and sort the mean survival rate by title
title_survival_rates = df.groupby('Title')['Survived'].mean().sort_values(ascending=False)

print(title_survival_rates)
```
